```
# Packages: numpy, opencv, matplotlib, scikit-image
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
# from mpl_toolkits.axes_grid1 import ImageGri
from skimage.color import rgb2gray
from skimage.io import imread
from skimage import filters
```

```
# Read original image in grayscale
image = rgb2gray(imread('chicken.jpg'))
```

```
fig, (ax, ax2, ax3, cax) = plt.subplots(ncols=4,figsize=(15,3), 
                  gridspec_kw={"width_ratios":[1,1,1,0.05]})
fig.subplots_adjust(wspace=0.3)

im  = ax.imshow(image)
ax.set_title('Original', fontsize=20)
ax.set_xticks([])
ax.set_yticks([])

sobelx = cv.Sobel(image, cv.CV_64F,1,0,ksize=5)
im2 = ax2.imshow(sobelx, vmin=0, vmax=1, cmap='gray')
ax2.set_title('Sobel_x', fontsize=20)
ax2.set_xticks([])
ax2.set_yticks([])

sobely = cv.Sobel(image, cv.CV_64F, 0, 1, ksize = 5)
im3= ax3.imshow(sobely,vmin=0, vmax=1, cmap = 'gray')
ax3.set_title('Sobel_y', fontsize=20)
ax3.set_xticks([])
ax3.set_yticks([])

from mpl_toolkits.axes_grid1.inset_locator import InsetPosition
ip = InsetPosition(ax3, [1.05,0,0.05,1]) 
cax.set_axes_locator(ip)
fig.colorbar(im, cax=cax, ax=[ax,ax3])

plt.show()
```
